```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# 加载数据
X, y = make_circles(n_samples=1000, noise=0.2, random_state=42)
print (X.shape)
print (y.shape)
```

```python
# 可视化
plt.figure(figsize=(12, 5))

#### 原数据（二维数据可以直接画）
# 2D散点图
plt.subplot(1, 2, 1)
plt.scatter(X[y==0, 0], X[y==0, 1], c='red', alpha=0.7, label='target=0')
plt.scatter(X[y==1, 0], X[y==1, 1], c='blue', alpha=0.7, label='target=1')
plt.xlabel(f'feature1 )')
plt.ylabel(f'feature2 )')
plt.title('make moon')
plt.legend()
plt.grid(True)


plt.show()
```
